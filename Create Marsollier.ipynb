```r
library(Seurat)
library(Signac)
library(SingleCellExperiment)
library(scran)
library(scater)
library(Matrix)
library(ChromSCape)
library(stringr)
```

```r
create_matrix <- function(cs_path, analysis) {
    datamatrix <- qs::qread(file.path(cs_path, 'ChromSCape_analyses', analysis, 'datamatrix.qs'))
    annot <- qs::qread(file.path(cs_path, 'ChromSCape_analyses', analysis,'annot_raw.qs'))
    sce <- create_scExp(datamatrix, annot)
    return(sce)
}

save_sce <- function(path, sce) {
    writeMM(counts(sce), file.path(path, 'matrix.mtx'))
    write(colnames(sce), file.path(path, 'barcodes.tsv'))
    write.table(data.frame(rownames(sce))[c('rownames.sce.', 'rownames.sce.')],
                file=file.path(path, 'bins.tsv'),
                row.names=FALSE,
                col.names=FALSE,
                quote=FALSE,
                sep="\t")
}
```

```r
for (x in c(5, 10, 20, 30, 50, 100, 200, 300, 500, 1000)) {
  analysis <- paste0('MarsolierK27.', x, 'k')
  sce <- create_matrix('/home/gamazeps/R/x86_64-pc-linux-gnu-library/4.2/ChromSCape', analysis)
  
  # Use proper granges
  rownames(sce) <- sub('_', '-', sub("_",":",rownames(sce)))

  # Filter cells
  cell_count <- colSums(counts(sce))
  sce <- sce[,which(cell_count >= 1000)]
  cell_count <- colSums(counts(sce))
  sce <- sce[,which(cell_count <= 5000)]
  # Filter features
  feature_count <- rowSums(counts(sce))
  sce <- sce[which(feature_count >= 3),]
  
  save_sce(paste0('/home/gamazeps/data/matrices/scChip_Marsolier_2022/H3K27me3/raw/', x, 'k'), sce)
}
```
